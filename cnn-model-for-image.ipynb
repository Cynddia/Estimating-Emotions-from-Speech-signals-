```python
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.metrics import MeanSquaredError
import cv2
```

```python
model = load_model('/kaggle/input/cnn-model-spectogram/tensorflow2/default/1/cnn_model.h5', 
                   custom_objects={'mse': MeanSquaredError})
```

```python
def preprocess_spectrogram(spectrogram_path):
    img = cv2.imread(spectrogram_path, cv2.IMREAD_GRAYSCALE)
    
    img_resized = cv2.resize(img, (128, 128))  # Resize to (128, 128)
    
    img_normalized = img_resized / 255.0
    
    img_expanded = np.expand_dims(img_normalized, axis=-1)  # Add channel dimension (1)
    img_expanded = np.expand_dims(img_expanded, axis=0)  # Add batch dimension (1)
    
    return img_expanded
```

```python
def predict_values(spectrogram_path):
    img_preprocessed = preprocess_spectrogram(spectrogram_path)
    
    predicted_values = model.predict(img_preprocessed)
    
    valence, arousal, dominance = predicted_values[0]
    
    return valence, arousal, dominance
```

```python
spectrogram_path = '/kaggle/input/trial-img/Ses04M_script01_3_M015.png' 
valence, arousal, dominance = predict_values(spectrogram_path)

print(f"Predicted Valence: {valence}")
print(f"Predicted Arousal: {arousal}")
print(f"Predicted Dominance: {dominance}")
```
